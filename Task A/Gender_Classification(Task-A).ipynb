```python
DATA_DIR = '/content/drive/MyDrive/Task_A'
```

```python
import os
from ultralytics import YOLO

model = YOLO("yolov8n-cls.pt")

results = model.train(data=DATA_DIR,epochs=30,imgsz=64)
```

```python
trained_model = YOLO('/content/runs/classify/train/weights/best.pt')
```

```python
from matplotlib import pyplot as plt
import cv2
def detectGender(image_path):
  results = trained_model.predict(image_path)
  for result in results:
        img = result.plot()  # Draw bounding boxes
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()
```

```python
detectGender('/content/drive/MyDrive/img11.jpg')
```
